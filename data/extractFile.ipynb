```python
import pandas as pd
```

```python
salaireMoyen = pd.read_excel('salaireMoyen.xlsx', index_col=0) 
```

```python
salaireMoyen.head()
```

```python
dfMoyen = pd.DataFrame(columns=['Pays', 'Année', 'SalaireMoyen'])

pays_unique = salaireMoyen.index.unique()
```

```python
for p in pays_unique:
    df2 = salaireMoyen[salaireMoyen.index == p]
    df2= df2.loc[df2['Devise'] == 'Devise: Dollar américain'].groupby('Année').mean()
    df2["Pays"] = p
    df2["Année"] = df2.index
    df2["SalaireMoyen"] = df2["Valeur"]
    dfMoyen = pd.concat([dfMoyen, df2[["Pays", "Année", "SalaireMoyen"]]])
```

```python
dfMoyen
```

```python
salaireMin = pd.read_excel('salaireMin.xlsx', index_col=0)
salaireMin.head()
```

```python
dfMin = pd.DataFrame(columns=['Pays', 'Année', 'SalaireMin'])
pays_unique_min = salaireMin.index.unique()
```

```python
for p in pays_unique_min:
    df2 = salaireMin[salaireMin.index == p]
    if (len(df2.index) > 1):
        df2 = df2.loc[df2['Unnamed: 2'] == 'Devise: Dollar américain'].groupby('Unnamed: 3').mean()
        if (len(df2.index) > 1):
            df2["Pays"] = p
            df2["Année"] = df2.index
            df2["SalaireMin"] = df2["Unnamed: 4"]
            dfMin = pd.concat([dfMin, df2[["Pays", "Année", "SalaireMin"]]])
```
